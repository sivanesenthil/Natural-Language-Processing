```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
# Download required NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
text = """
Artificial intelligence (AI) and machine learning (ML) are two related but distinct fields in 
the realm of computer science. AI refers to a computer system's ability to perform tasks that typically require 
human intelligence to complete, such as recognizing speech, making decisions, and solving problems. On the other hand, 
ML is a subset of AI that focuses on enabling a computer system to learn from data and improve its performance on a 
specific task over time without being explicitly programmed. In layman's terms, AI makes machines capable of performing
tasks like humans, while ML helps machines improve their performance in completing specific tasks without being explicitly 
programmed.
"""

# Step 1: Text Processing
# Tokenize the text
words = word_tokenize(text.lower())
```

```python
print(len(words))
```

```python
words
```

```python
#Empty list to store words
words_no_punc= []
```

```python
#To Remove punctuation marks
for w in words:
    if w.isalpha():
        words_no_punc.append(w.lower())
words_no_punc
```

```python
print
(len(words_no_punc))
```

```python
from nltk.corpus import stopwords
#List stopwords
stopwords = set(stopwords.words('english'))
print(stopwords)
```

```python
#Empty list to store words
new_words=[]
for w in words_no_punc:
    if w not in stopwords:
        new_words.append(w)
print(new_words)
```

```python
from nltk.probability import FreqDist
fdist = FreqDist(new_words)
fdist.most_common(10)
```

```python
#Library
from wordcloud import WordCloud
#Library to plot the wordcloud
import matplotlib.pyplot as plt
#Generating the worcloud
wordcloud = WordCloud().generate_from_frequencies(fdist)
#Plot the wordcloud
plt.figure(figsize = (8,8))
plt.imshow(wordcloud)
```

```python
# Create circular mask
x, y = np.ogrid[:800, :800]
mask = (x - 400) ** 2 + (y - 400) ** 2 > 390 ** 2
mask = 255 * mask.astype(int)

# Generate and plot the word cloud
wordcloud = WordCloud(width=800, height=800, background_color='white', mask=mask).generate_from_frequencies(fdist)

plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
# Create triangular mask
x, y = np.ogrid[:800, :800]
mask = (y < 1.5 * x) & (y < -1.5 * x + 1200)
mask = 255 * mask.astype(int)
# Generate and plot the word cloud
wordcloud = WordCloud(width=800, height=800, background_color='white', mask=mask, contour_width=3, contour_color='black').generate_from_frequencies(fdist)

plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
# Create triangular mask
x, y = np.ogrid[:800, :800]
triangle_height = 800
triangle_base = 800
mask = (y >= (triangle_height / triangle_base) * x) & (y >= (triangle_height / triangle_base) * (triangle_base - x))
mask = 255 * mask.astype(int)

# Generate and plot the word cloud
wordcloud = WordCloud(width=800, height=800, background_color='white', mask=mask, contour_width=3, contour_color='black').generate_from_frequencies(fdist)

plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
